## 이미지

> 이미지를 여러 패턴으로 분류해서, 패턴형 데이터를 학습하여 새로운 이미지가 주어질 때 이것이 무엇인지 정확하게 알아내는 원리이다. 주로 딥러닝 혹은 신경망네트워크가 이 기능을 구현하는 데 쓰인다.

1. 내가 원하는 사진을 검색해주는 구글 포토(Google Photo) 서비스
> 이미지 인식 기술은 이미 우리 생활 가까이에서 많이 사용되고 있습니다. 스마트폰으로 사진 촬영을 자주 하시는 분이라면 구글 포토 서비스에 사진을 백업해두고 있을 텐데요. 서비스 이용을 안 하신 다면 지금이라도 신청하시길 권해드립니다. 고화질 원본 이미지가 아니라면 무제한 저장 용량을 제공하거든요. 그리고, 거기에 덧붙여 구글 포토는 놀라운 분류와 검색 기능을 가지고 있습니다.

2. 딥페이크(Deepfake)
> 러닝(Deep Learning)과 가짜를 의미하는 페이크(fake)의 합성어인 이 용어는 실제 영상처럼 가공된 가짜 영상을 의미합니다. 바로 인공지능화된 학습을 통해 실제 영상과 구분하기 어려운 영상을 만들어 내는 기술인데요.딥페이크 기술은 유명인들을 대상으로 악의적인 영상이 만들어져 유포되고, 가짜 뉴스를 만들어 시민들에게 혼란을 야기하며 정치적인 목적으로 악용되기도 합니다.

3. **이미지 분석을 통해 질병을 예측하고 분석하는 의료 서비스(관심있는 기술)**
> IBM Watson의 의학연구부분을 시작으로 영상의학 판독 분야의 임상의사지원시스템을 포함하여 정형/비정형 임상 데이터를 분석하여 질병 진단과 예측의 정확도를 높이는 방향으로 진화했습니다.
암 환자 진단, 영상 의료 데이터 분석 및 판독, 유전체 분석, 임상 시험 등에서 활용하고 있으며, 이외에도 심전도, 혈당, 혈압 등의 연속적인 생체 데이터를 분석하여 위험 징후를 조기에 예측(Predictive), 예방(Preventive)하는 분야에 활용

+ 목표 및 견해
> **우리나라 성인 남녀 사망률 상위를 차지하는 질병, 폐암은 말기에 발견되는 경우 생존율 6.1%지만 조기 발견의 경우 생존율이 64%까지 상승한다고 합니다. 그런데 국내 폐암 환자 조기진단 비율은 20.7%에 불과합니다. 하지만 인공지능과 접목하여 이상부분을 손쉽게 찾아 의사의 판독을 보조하는 이러한 기술이야말로 현재 의료업계의 혁신적인 기술이라 할수있다고 생각합니다.**

##자율주행

> 자율주행은 운전자 개입 없이 자동차가 스스로 도로 환경을 인식하고 조항, 변속 장치 등을 제어하여 목적지까지 주행하는 것을 가리킵니다.

1. 안전과 편의성을 돕는 선택적 능동 제어 단계
> "순항 제어”를 뜻하는 크루즈 컨트롤은 고속도로에서 흔히 사용되는 기능입니다. 설정된 속도로 일정하게 주행이 가능하여 운전자의 편의를 높여줍니다.  국내에서는 1985년에 발매된 현대 쏘나타에서 크루즈 컨트롤이 최초로 장착되었고, 현재는 보편적으로 적용되고 있는 기능입니다..
차선 유지 지원 시스템, 충돌 위험시 자동 브레이크 제어 기술 등이 있답니다. 

2. 통합 능동제어 단계
> 주행 조향 보조 시스템은 방향 지시등을 켜지 않고 자동차가 차선을 이탈하게 되면, 조향장치가 인식하여 자동으로 복귀시켜주는 기술입니다.속도로 주행지원 시스템 HDA은 기존의 차선유지기능뿐만 아니라, 다른 차가 앞에 서있으면 자동으로 멈춘 후 다시 출발하는 차간거리제어기능, 제한 속도 구간에서는 운전자의 개입없이도 스스로 속도 조절이 가능한 똑똑한 기능들이 포함되어 있습니다.

3. **제한적 자율주행 단계(관심있는 기술)**
> 자동차가 도로정보를 스스로 해석하여 방해물을 피하거나 막히는 구간을 우회하여 지나가기도 하는 단계입니다. 하지만 아주 급박한 상황에서는 운전자의 조작이 불가피한 단계이기도 합니다.

* 목표 및 견해
> **'사람이 자동차 운전을 하는 것이 불법인 시대가 가까운 미래에 올 것이다' 테슬라 최고경영자가 언급한 자율주행차의 미래입니다. 최근의 자율주행차 발전속도는 어느때보다 빠릅니다.이러한 시대에 자율주행차에 발전에 되는 그러한 인공지능 기술을 많이 배우고 싶습니다.**
